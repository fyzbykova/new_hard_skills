### Matplotlib

```python
# Корректное отображение графиков в юпитере

%matplotlib inline

# Все функции Matplotlib API, в частности plot и close, находятся в модуле matplotlib.pyplot

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221) # кодируем графики (строки, столбы, порядок)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

plt.show()
plt.close()
```

```python
from numpy.random import randn
```

```python
fig = plt.figure(figsize = (14, 5))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(122)
ax3 = fig.add_subplot(223)

# plt.plot(randn(100).cumsum(), 'k --')

# Метод fig.add_subplot возвращает объект AxesSubplot, что позволяет рисовать в любом подграфике, 
# вызывая методы этого объекта:

# Гистограмма 
f = ax1.hist(randn(100), bins=20, color='y', alpha=0.5)
# Скэтерплот
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
# Просто линия
ax3.plot(randn(100).cumsum(), 'k--')
```

```python
f
```

```python
# Отдельно данный метод тоже будет работать, так как при необходимости он автоматически
# создаёт рисунок и подграфик

plt.plot(randn(100).cumsum())
```

```python
fig, axes = plt.subplots(2, 3)
fig.set_figheight(5)
fig.set_figwidth(15)

axes[0, 1].scatter(np.arange(30), np.arange(30) + 3 * randn(30))
axes[1, 1].bar(np.arange(10), randn(10))

plt.show()
```

```python
axes
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(15)

for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins=50, color='r', alpha=0.5)

# Достаточно популярный способ задания пустого пространства вокруг графиков

plt.subplots_adjust(wspace=0.0, hspace=0.0)
```

```python
fig, axes = plt.subplots(2, 2, sharex=False, sharey=False)
fig.set_figheight(5)
fig.set_figwidth(15)

for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins=50, color='r', alpha=0.5)

# Достаточно популярный способ задания пустого пространства вокруг графиков

plt.subplots_adjust(wspace=0.2, hspace=0.2)
```

```python
# Сохранение рисунка

plt.savefig('fig1.png', dpi=400, bbox_inches='tight')
```
